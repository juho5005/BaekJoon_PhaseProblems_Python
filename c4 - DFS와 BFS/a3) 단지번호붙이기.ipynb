# DFS 풀이

```python
# 단지번호붙이기
# 1 : 집이 있는 곳
# 0 : 집이 없는 곳

# DFS 
import sys 
n = int(sys.stdin.readline()) # 5 <= N <= 25
sys.setrecursionlimit(10**6)

graph = [
    list(sys.stdin.readline().strip())
    for _ in range(n)
]

visited = [
    [False for _ in range(n)]
    for _ in range(n)
]

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<n

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] :
        return False 
    if graph[x][y] == '0' :
        return False 
    return True 

# DFS 
def dfs(x, y) :
    global house_cnt
    house_cnt += 1
    for dx, dy in zip(dxs, dys) :
        nx, ny = x+dx, y+dy 
        if can_go(nx, ny) :
            visited[nx][ny] = True 
            dfs(nx, ny)

house_lst = []
cnt = 0
for i in range(n) :
    for j in range(n) :
        if not visited[i][j] and graph[i][j] == '1' :
            house_cnt = 0
            visited[i][j] = True 
            dfs(i, j)
            cnt +=1 
            house_lst.append(house_cnt)

print(cnt)
house_lst.sort()
for elem in house_lst :
    print(elem)
```

# BFS 풀이 

```python
# 단지번호붙이기
# 1 : 집이 있는 곳
# 0 : 집이 없는 곳

# BFS 
from collections import deque 
import sys 
n = int(sys.stdin.readline()) # 5 <= N <= 25
q = deque()

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

graph = [
    list(sys.stdin.readline().strip())
    for _ in range(n)
]

visited = [
    [False for _ in range(n)]
    for _ in range(n)
]

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<n

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] :
        return False 
    if graph[x][y] == '0' :
        return False 
    return True 

# BFS 
def bfs() :
    global house_cnt
    while q : 
        x, y = q.popleft()
        for dx, dy in zip(dxs, dys) :
            nx, ny = x+dx, y+dy 

            if can_go(nx, ny) :
                visited[nx][ny] = True 
                q.append((nx, ny))
                house_cnt += 1
                

house_lst = []
cnt = 0
for i in range(n) :
    for j in range(n) :
        if not visited[i][j] and graph[i][j] == '1' :
            house_cnt = 1
            visited[i][j] = True 
            q.append((i, j))
            bfs()
            cnt +=1 
            house_lst.append(house_cnt)

print(cnt)
house_lst.sort()
for elem in house_lst :
    print(elem)
```
