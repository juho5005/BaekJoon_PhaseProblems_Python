### cnt를 모든 함수에 돌리지 않고, 마지막에 visited의 함수에서 1을 제외한 True의 개수를 세주는 것이 시간의 효율이 더 좋았다.

# DFS ( 인접행렬 방식 )

```python
# 바이러스
import sys 

n = int(sys.stdin.readline()) # N <= 100 O(N^4)
c = int(sys.stdin.readline()) # 간선의 수

# 바이러스에 걸리게 되는 컴퓨터의 수
cnt = 0

graph = [
    [0 for _ in range(n+1)]
    for _ in range(n+1)
]

for _ in range(c) :
    start, end = tuple(map(int, sys.stdin.readline().split()))
    graph[start][end] = 1
    graph[end][start] = 1

visited = [
    False 
    for _ in range(n+1)
]

def dfs(vertex) :
    global cnt
    for curr_v in range(1, n+1) :
        if graph[vertex][curr_v] and not visited[curr_v] :
            cnt += 1
            visited[curr_v] = True 
            dfs(curr_v)

root_vertex = 1
visited[root_vertex] = True 
dfs(root_vertex)

print(cnt)
```

# BFS ( 인접행렬 방식 )

```python
# 바이러스
import sys 
from collections import deque 

n = int(sys.stdin.readline()) # N <= 100 O(N^4)
c = int(sys.stdin.readline()) # 간선의 수

# 바이러스에 걸리게 되는 컴퓨터의 수
cnt = 0

# 덱 선언
q = deque()

graph = [
    [0 for _ in range(n+1)]
    for _ in range(n+1)
]

for _ in range(c) :
    start, end= tuple(map(int, sys.stdin.readline().split()))
    graph[start][end] = 1
    graph[end][start] = 1

visited = [
    False 
    for _ in range(n+1)
]

def bfs() :
    global cnt
    while q :
        curr_v = q.popleft()
        for next_v in range(1, n+1) :
            if graph[curr_v][next_v] and not visited[next_v] :
                visited[next_v] = True 
                q.append(next_v)
                cnt += 1

root_vertex = 1
visited[root_vertex] = True 
q.append(root_vertex)
bfs()

print(cnt)
```

# DFS ( 인접 리스트를 이용한 방식 )

```python
# 바이러스
import sys 

n = int(sys.stdin.readline()) # N <= 100 O(N^4)
c = int(sys.stdin.readline()) # 간선의 수
cnt = 0

graph = [
    []
    for _ in range(n+1)
]

visited = [
    False 
    for _ in range(n+1)
]

for _ in range(c) :
    start, end = tuple(map(int, sys.stdin.readline().split()))
    graph[start].append(end)
    graph[end].append(start)

for elem in graph :
    elem.sort()

def dfs(curr_v) :
    global cnt 
    
    for next_v in graph[curr_v] :
        if not visited[next_v] :
            cnt += 1
            visited[next_v] = True 
            dfs(next_v)

root_vertex = 1
visited[root_vertex] = True 
dfs(root_vertex)

print(cnt)
```
