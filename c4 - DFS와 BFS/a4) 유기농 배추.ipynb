# DFS 풀이 

```python
# 유기농 배추
# 배추가 모여있는 곳에는 배추지렁이가 한 마리만 존재하면 된다.
# 배추가 심어져 있는 땅(1), 배추가 없는 땅(0)
# 주어진 테스트 케이스에서 배추지렁이는 몇 마리가 필요한가?

# DFS

import sys 
t = int(sys.stdin.readline())
sys.setrecursionlimit(10**6)

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<m 

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] :
        return False 
    if graph[x][y] == 0 :
        return False 
    return True 

def dfs(x, y) :
    for dx, dy in zip(dxs, dys) :
        nx, ny = x+dx, y+dy 

        if can_go(nx, ny) :
            visited[nx][ny] = True 
            dfs(nx, ny)

for _ in range(t) :
    # 배추밭 가로(M), 세로(N), 배추가 심어져있는 위치의 개수(K), 위치가 같은 경우 X
    m, n, k = tuple(map(int, sys.stdin.readline().split()))
    
    graph = [
        [0 for _ in range(m)] 
        for _ in range(n)
    ]

    visited = [
        [False for _ in range(m)]
        for _ in range(n)
    ]

    for _ in range(k) :
        x, y = tuple(map(int, sys.stdin.readline().split()))
        graph[y][x] = 1 
    
    cnt = 0
    for i in range(n) :
        for j in range(m) :
           if not visited[i][j] and graph[i][j] == 1 :
               visited[i][j] = True 
               dfs(i, j)
               cnt += 1
    print(cnt)
```

# BFS 풀이 

```python
# 유기농 배추
# 배추가 모여있는 곳에는 배추지렁이가 한 마리만 존재하면 된다.
# 배추가 심어져 있는 땅(1), 배추가 없는 땅(0)
# 주어진 테스트 케이스에서 배추지렁이는 몇 마리가 필요한가?

# BFS

from collections import deque 
import sys 
t = int(sys.stdin.readline())
q = deque()

dxs = [-1, 1, 0, 0]
dys = [0, 0, -1, 1]

def in_range(x, y) :
    return x>=0 and x<n and y>=0 and y<m 

def can_go(x, y) :
    if not in_range(x, y) :
        return False 
    if visited[x][y] :
        return False 
    if graph[x][y] == 0 :
        return False 
    return True 

def bfs() :
    while q :
        x, y = q.popleft()
        for dx, dy in zip(dxs, dys) :
            nx, ny = x+dx, y+dy 
            
            if can_go(nx, ny) :
                visited[nx][ny] = True 
                q.append((nx, ny))

for _ in range(t) :
    # 배추밭 가로(M), 세로(N), 배추가 심어져있는 위치의 개수(K), 위치가 같은 경우 X
    m, n, k = tuple(map(int, sys.stdin.readline().split()))
    
    graph = [
        [0 for _ in range(m)] 
        for _ in range(n)
    ]

    visited = [
        [False for _ in range(m)]
        for _ in range(n)
    ]

    for _ in range(k) :
        x, y = tuple(map(int, sys.stdin.readline().split()))
        graph[y][x] = 1 
    
    cnt = 0
    for i in range(n) :
        for j in range(m) :
           if not visited[i][j] and graph[i][j] == 1 :
               visited[i][j] = True 
               q.append((i, j))
               bfs()
               cnt += 1
    print(cnt)
```
