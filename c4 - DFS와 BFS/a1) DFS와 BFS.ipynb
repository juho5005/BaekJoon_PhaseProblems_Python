## 인접 행렬 방식으로 풀었을 때

```python
from collections import deque 
import sys 
n, m, y = tuple(map(int, sys.stdin.readline().split()))
q = deque()


# 양방향 간선
linked_vertex = [
    tuple(map(int, sys.stdin.readline().split()))
    for _ in range(m)
]

# 1) 인접행렬 방식으로 해결할 때 (DFS, BFS 공통적용)
graph = [
    [0 for _ in range(n+1)]
    for _ in range(n+1)
]

dfs_visited = [
    False 
    for _ in range(n+1)
]

bfs_visited = [
    False 
    for _ in range(n+1)
]

# 2) graph의 값 초기화 단계
def initial() :
    for start, end in linked_vertex :
        graph[start][end] = 1
        graph[end][start] = 1 # 양방향 그래프 이므로 대칭되게 1로 초기화

initial()

def dfs(vertex) :
    for curr_v in range(1, n+1) :
        if graph[vertex][curr_v] and not dfs_visited[curr_v] :
            dfs_path.append(curr_v)
            dfs_visited[curr_v] = True 
            dfs(curr_v) 

def bfs() :
    while q :
        curr_v = q.popleft()
        for next_v in range(1, n+1) :
            if graph[curr_v][next_v] and not bfs_visited[next_v] :
                bfs_path.append(next_v)
                bfs_visited[next_v] = True
                q.append(next_v)

# 3) 마지막 부분 
dfs_path = [y]
bfs_path = [y]

root_vertex = y
dfs_visited[root_vertex] = True 
bfs_visited[root_vertex] = True 

dfs(root_vertex)

q.append(root_vertex)
bfs()


print(*dfs_path)
print(*bfs_path)
```

## 연결리스트로 풀었을 때
## 주의사항) 
## 주어지는 간선이 연결하는 두 정점의 번호를 정렬해야한다!
#### 정점 번호가 작은 것부터 방문하기 때문이다.

```python
from collections import deque 
import sys 
n, m, y = tuple(map(int, sys.stdin.readline().split()))
q = deque()

# 연결리스트로 해결할 때
graph = [
    []
    for _ in range(n+1)
]


# 양방향 간선
for _ in range(m) :
    x1, x2 = tuple(map(int, sys.stdin.readline().split()))

    graph[x1].append(x2)
    graph[x2].append(x1)

for g in graph : 
    g.sort() 

dfs_visited = [
    False
    for _ in range(n+1)
]

bfs_visited = [
    False 
    for _ in range(n+1)
]

def dfs(vertex) :
    for curr_v in graph[vertex] :
        if not dfs_visited[curr_v] :
            dfs_path.append(curr_v)
            dfs_visited[curr_v] = True 
            dfs(curr_v)

def bfs() :
    while q :
        curr_v = q.popleft()
        for next_v in graph[curr_v] :
            if not bfs_visited[next_v] :
                bfs_path.append(next_v)
                bfs_visited[next_v] = True 
                q.append(next_v) 

root_vertex = y 

dfs_path = [y]
bfs_path = [y]

dfs_visited[root_vertex] = True 
bfs_visited[root_vertex] = True 

dfs(root_vertex) 

q.append(root_vertex)
bfs()

print(*dfs_path)
print(*bfs_path)
```
